```python
file = open('hamlet.txt', encoding ='utf-8')
text = file.read()
```

```python
text
```

```python
# Tokenize data: Handling punctuations and lowercasing the text 
from gensim.utils import simple_preprocess 

# preprocess the file to get a list of tokens 
token_list =[] 

for sentence in text.split('.'):   
    # the simple_preprocess function returns a list of each sentence 
    token_list.append(simple_preprocess(sentence, deacc = True))

print (token_list[:5])
```

```python
# Import gensim corpora
from gensim import corpora 

# storing the extracted tokens into the dictionary 
my_dictionary = corpora.Dictionary(token_list) 

# print the dictionary 
print(my_dictionary)
```

```python
# save your dictionary to disk 
my_dictionary.save('dictionary.dict')

# load back 
load_dict = corpora.Dictionary.load('dictionary.dict')

print(load_dict)
```

# text preprocessing use pattern

```python
from pattern.en import parse
from pattern.en import pprint
pprint(parse('Hello I am John, I work at the bank.',relations=True, lemmata=True))
```

```python
pprint(parse('Enzo Ferrari was not initially interested in the idea of producing road cars when he formed Scuderia Ferrari in 1929, with headquarters in Modena.', relations  = True,tokenize= True, lemmata= True))
```

```python
from pattern.en import ngrams
#n grams
print(ngrams("There is nothing either good or bad, but thinking makes it so.", n=3))
```

```python
from pattern.en import sentiment
print(sentiment("The course of true love never did run smooth."))
```

```python
from pattern.en import suggest
print(suggest("Aerplane"))
```

```python
from pattern.en import quantify
a = quantify(['Car', 'Car', 'Bus', 'Truck', 'Train', 'Car', 'Bus', 'Truck','ship'])
print(a)
```

```python
from pattern.en import number, numerals
print(number("ten thousand five hundred seventy one"))
print(numerals(78567.79585, round=3))
```

```python
from pattern.en import pluralize, singularize
print(pluralize('car'))
print(singularize('BUSES'))
```

```python
from pattern.en import comparative, superlative
print(comparative('bad'))
print(superlative('bad'))
```
