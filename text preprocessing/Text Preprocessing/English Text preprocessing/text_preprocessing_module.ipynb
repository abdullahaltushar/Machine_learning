html tag remove

```python
def html_tag_remover(text):
    soup=BeautifulSoup(text,"html.parse")
    return soup.get_text()
```

Punctuation remove

```python
def remove_punctuation(text):
    return re.sub('r[^\w\s]','',text)
```

Remove \n tag

```python
def replace_new_line(text):
    return re.sub(r'(\n)', ' ', text)
```

Remove underline tag

```python
def replace_under_line(text):
    return re.sub(r'_', '', text)
```

Remove extra space

```python
def replace_space(text):
    return re.sub(r' +',' ',text)
```

Replace contractions in string of text

```python
def contraction(text):
    return contractions.fix(text)
```

Word split

```python
import wordninja
def word_split(text):
    return wordninja.split(text)
```

tokenization data

```python
def tokenize(text):
    return nltk.word_tokenize(text)
```

Non Ascii Remover

```python
import re, string, unicodedata
def remove_non_ascii(text):
    new_words = []
    for word in text:
        new_word = unicodedata.normalize('NFKD', word).encode('ascii', 'ignore').decode('utf-8', 'ignore')
        new_words.append(new_word)
    return new_words
```

Convert Lower case

```python
def to_lowercase(text):
    new_words = []
    for word in text:
        new_word = word.lower()
        new_words.append(new_word)
    return new_words
```

Replace digits into word

```python
def replace_numbers(text):
    p = inflect.engine()
    new_words = []
    for word in text:
        if word.isdigit():
            new_word = p.number_to_words(word)
            new_words.append(new_word)
        else:
            new_words.append(word)
    return new_words
```

Remove stopwords

```python
def remove_stopwords(text):
    new_words = []
    for word in text:
        if word not in stopwords.words('english'):
            new_words.append(word)
    return new_words
```

Stemming 

```python
from nltk.stem import PorterStemmer
def stemmer(text):
    stemmer= PorterStemmer()
    stemmers=[]
    for word in text:
        stemm= stemmer.stem(word)
        stemmers.append(stemm)

    return stemmers
```

lemmatization

```python
def lemmatize_verbs(text):
    """Lemmatize verbs in list of tokenized words"""
    lemmatizer = WordNetLemmatizer()
    lemmas = []
    for word in text:
        lemma = lemmatizer.lemmatize(word, pos='v')
        lemmas.append(lemma)
    return lemmas
```

part of speech 

```python
def pos_tag(text):
    return nltk.pos_tag(text)
```

called needed module

```python
def preprocess(text):
#     #text=html_tag_remover(text)
    text=remove_punctuation(text)
    text=replace_new_line(text)
    text=replace_under_line(text)
    text=replace_space(text)
    text=contraction(text)
    words=word_split(text)
    #text=tokenize(text)
    word=remove_non_ascii(words)
    word_lower=to_lowercase(word)
    word_string= replace_numbers(word_lower)
    #text=stemmer(text)
    word_stop=remove_stopwords(word_string)
    word_lem=lemmatize_verbs(word_stop)
    word_final= pos_tag(word_lem)
    return word_final
```

```python
word_com = preprocess(text)
print(word_com)
```
