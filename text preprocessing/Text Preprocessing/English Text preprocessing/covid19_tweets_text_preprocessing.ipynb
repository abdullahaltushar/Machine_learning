# remove url link

```python
df["clean_text"] = df["text"].apply(lambda s: ' '.join(re.sub("https:\/\/.*|http:\/\/.*", " ", s).split()))
df[['text','clean_text']].iloc[94807]
```

```python
df
```

# Remove punctuations

```python
df["clean_text"] = df["clean_text"].apply(lambda s: ' '.join(re.sub('[.,!?:;-=..."@#_]', " ", s).split()))
df[["text","clean_text"]]
```

# remove digits 

```python
df["clean_text"].replace('\d+', '', regex=True, inplace=True)
df[["text","clean_text"]]
```

# remove ascii 

```python
def deEmojify(inputString):
    return inputString.encode('ascii', 'ignore').decode('ascii')
df["clean_text"] = df["clean_text"].apply(lambda s: deEmojify(s))
```

```python
df["clean_text"]
```

```python
import nltk
stopwords= nltk.corpus.stopwords.words("english")
stopwords
```

```python
def tokenization(text):
    token=text.split()
    return token
df["clean_text"]=df["clean_text"].apply(lambda x: tokenization(x))
df
```

```python
def remove_stop_words(text):
    data=[i for i in text if i not in stopwords]
    return data   
```

```python
df["clean_text"]=df["clean_text"].apply(lambda x: remove_stop_words(x))
df["clean_text"]
```

```python
from nltk.stem import WordNetLemmatizer
lemmatiser = WordNetLemmatizer()
df["clean_text"] = df["clean_text"].apply(lambda tokens: [lemmatiser.lemmatize(token, pos='v') for token in tokens])
df[["text","clean_text"]]
```

# tokenization

```python
#importing libraries
import spacy

#instantiating English module
nlp = spacy.load("en_core_web_sm")
#sample
x = "Embracing and analyzing self failures (of however multitude) is a virtue of nobelmen."

#creating doc object containing our token features
doc = nlp(x)

#Creating and updating our list of tokens using list comprehension 
tokens = [token.text for token in doc]
print(tokens)
```

# lemmatization

```python
#importing libraries
import spacy

#instantiating English module
nlp = spacy.load("en_core_web_sm")

#sample
x = "Running down the street with my best buddy."

#creating doc object containing our token features
doc = nlp(x)

#Creating and updating our list of tokens using list comprehension 
tokens = [[token.text,token.lemma_] for token in doc]
print(tokens)
```

```python
#importing libraries

import spacy

#instantiating English module

nlp = spacy.load("en_core_web_sm")

#sample
x = "Running down the street with my best buddy."

#creation of doc object containing our token features

doc = nlp(x)

#Creating and updating our list of tokens using list comprehension 

tokens = [token.text for token in doc]

print(tokens)

#Creating and updating our list of filtered tokens using list comprehension 

filtered = [token.text for token in doc if token.is_stop == False]

print(filtered)
```

```python
#sample 
x = "BLIMEY!! Such an exhausting day, I can't even describe."

#creation of doc object containing our token features
doc = nlp(x)

#Unfiltered tokens 
tokens = [token.text for token in doc]
print(tokens)

#Filtering our tokens
filtered = [token.text for token in doc if token.is_stop == False and       
token.text.isalpha() == True]

print(filtered)
```

```python
import nltk
from nltk.tokenize import word_tokenize  as wt
from nltk import pos_tag as pt
nltk.download('averaged_perceptron_tagger')
text = 'I love my black little pug. His name is blacky. He loves to run'
token = wt(text) 
gm = "NP: {<DT>?<JJ>*<NN>}"
pos = pt(token) 
cp = nltk.RegexpParser(gm) 
tree = cp.parse(pos) 
```

# stemming different module

```python
from nltk.stem import PorterStemmer
porter = PorterStemmer()
words = ['eating','eats','eaten','puts','putting']
for word in words:
    print(word,"--->",porter.stem(word))
```

```python
from nltk.stem import SnowballStemmer
snowball = SnowballStemmer(language='english')
words = ['eating','eats','eaten','puts','putting']
for word in words:
    print(word,"--->",snowball.stem(word))
```

```python
from nltk.stem import LancasterStemmer
lancaster = LancasterStemmer()
words = ['eating','eats','eaten','puts','putting']
for word in words:
    print(word,"--->",lancaster.stem(word))
```
